```python
y=dataset["Class"].values
```

```python
x=dataset.copy()
```

```python
x.drop(["Class"],axis=1,inplace=True)
```

```python
x.head()
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
```

```python
x=scaler.fit_transform(x)
```

```python
x
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import confusion_matrix
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
```

```python
lr=LogisticRegression()
score_lr=cross_val_score(lr,x,y,cv=10,scoring="accuracy")
plt.plot(range(1,11),score_lr)
plt.xlabel("ith cross validation")
plt.ylabel("cross_val_score")
plt.title("Logistic Regression cross validation score")
print("mean_score->",score_lr.mean())
```

```python
knn=KNeighborsClassifier(n_neighbors=5)
score_knn=cross_val_score(knn,x,y,cv=10,scoring="accuracy")
plt.plot(range(1,11),score_knn)
plt.xlabel("ith cross validation")
plt.ylabel("cross_val_score")
plt.title("KNN cross validation score")
print("mean_score->",score_knn.mean())
```

```python
rfc=RandomForestClassifier(n_estimators=100)
score_rfc=cross_val_score(rfc,x,y,cv=10,scoring="accuracy")
plt.plot(range(1,11),score_rfc)
plt.xlabel("ith cross validation")
plt.ylabel("cross_val_score")
plt.title("RFC cross validation score")
print("mean_score->",score_knn.mean())
```

```python
xgb=XGBClassifier(n_estimators=100)
score_xgb=cross_val_score(xgb,x,y,cv=10,scoring="accuracy")
plt.plot(range(1,11),score_xgb)
plt.xlabel("ith cross validation")
plt.ylabel("cross_val_score")
plt.title("RFC cross validation score")
print("mean_score->",score_xgb.mean())
```

```python
#Applying PCA
from sklearn.decomposition import PCA
pca=PCA(n_components=5)
x=pca.fit_transform(x)
explained_variance=pca.explained_variance_ratio_
print(explained_variance)
```

```python
lr=LogisticRegression()
score_lr=cross_val_score(lr,x,y,cv=10,scoring="accuracy")
plt.plot(range(1,11),score_lr)
plt.xlabel("ith cross validation")
plt.ylabel("cross_val_score")
plt.title("Logistic Regression cross validation score after PCA")
print("mean_score->",score_lr.mean())
```

```python
xgb=XGBClassifier(n_estimators=100)
score_xgb=cross_val_score(xgb,x,y,cv=10,scoring="accuracy")
plt.plot(range(1,11),score_xgb)
plt.xlabel("ith cross validation")
plt.ylabel("cross_val_score")
plt.title("RFC cross validation score after PCA")
print("mean_score->",score_xgb.mean())
```

```python
knn=KNeighborsClassifier(n_neighbors=5)
score_knn=cross_val_score(knn,x,y,cv=10,scoring="accuracy")
plt.plot(range(1,11),score_knn)
plt.xlabel("ith cross validation")
plt.ylabel("cross_val_score")
plt.title("KNN cross validation score after PCA")
print("mean_score->",score_knn.mean())
```

```python
rfc=RandomForestClassifier(n_estimators=100)
score_rfc=cross_val_score(rfc,x,y,cv=10,scoring="accuracy")
plt.plot(range(1,11),score_rfc)
plt.xlabel("ith cross validation")
plt.ylabel("cross_val_score")
plt.title("RFC cross validation score after PCA")
print("mean_score->",score_knn.mean())
```
